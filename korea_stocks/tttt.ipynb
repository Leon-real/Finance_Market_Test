```python
import pandas as pd
import numpy as np
from pykrx import stock
import time
from datetime import datetime, timedelta
import os

pd.set_option('mode.chained_assignment',  None) # 경고 off
```

```python
### 날짜 업데이트
def Update_Stocks_OHCVL(File_name): # File_name의 양식은 "code_name stocks.xlsx"
    stocks = '000020_동화약품 stocks.xlsx'

    # 파일 있는지 확인하기
    if os.path.exists(os.path.join(os.getcwd(),'stocks lists',f'{stocks}')):
        print("파일 존재")
        code = stocks[:6]
        # 저장된 DataFrame 읽기
        df = pd.read_excel(os.path.join(os.getcwd(),'stocks lists',f'{stocks}'),index_col=0)
        pre_date = df.index[-1] # 저장된 엑셀의 마지막 시간 확인
        n_date = datetime.fromtimestamp(time.time()) # 현재 시간

        today_date = datetime.today().strftime("%Y%m%d") # 오늘 날짜

        if (n_date - pre_date).days >0:
            diff = (n_date - pre_date).days
            before_date = (datetime.today() - timedelta(diff)).strftime("%Y%m%d")
            add_df = stock.get_market_ohlcv_by_date(before_date,today_date,code)
            new_df = pd.concat([df[:-(len(add_df)-1)],add_df], ignore_index=False)
            df = new_df
            
    else: # 파일이 없을 경우 90일가량의 OHCVL 값을 가져와서 저장한다.
        today_date = datetime.today().strftime("%Y%m%d") # 오늘 날짜
        before_date = (datetime.today() - timedelta(90)).strftime("%Y%m%d") # 90 일 전의 날짜
        df = stock.get_market_ohlcv_by_date(before_date, today_date, code)
        df.to_excel(f"./stocks lists/{code}_{name} stocks.xlsx")
        
    return df
```
