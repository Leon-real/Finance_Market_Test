```python
import pandas as pd
import numpy as np
from pykrx import stock
import time
from datetime import datetime, timedelta
import os

pd.set_option('mode.chained_assignment',  None) # 경고 off

### 날짜 업데이트
def Update_Stocks_OHCVL(File_name):
    i = File_name
    stocks = i.split("_")
    code = stocks[0]
    name = stocks[1]
    
    # 파일 있는지 확인하기
    if os.path.exists(os.path.join(os.getcwd(),'stocks lists',f'{i}')):
        print(f"File Exist : {name}")
        # 저장된 DataFrame 읽기
        df = pd.read_excel(os.path.join(os.getcwd(),'stocks lists',f'{i}'),index_col=0)
        pre_date = df.index[-1] # 저장된 엑셀의 마지막 시간 확인
        n_date = datetime.fromtimestamp(time.time()) # 현재 시간

        today_date = datetime.today().strftime("%Y%m%d") # 오늘 날짜

        if (n_date - pre_date).days >0:
            diff = (n_date - pre_date).days
            before_date = (datetime.today() - timedelta(diff)).strftime("%Y%m%d")
            add_df = stock.get_market_ohlcv_by_date(before_date,today_date,code)
            new_df = pd.concat([df[:-(len(add_df)-1)],add_df], ignore_index=False)
            df = new_df
            print(f"Update File : {name}")
        
    return df

#*------------------------Main----------------------------*

# "stocks lists"라는 폴더에서 xlsx파일 가져오기
path = os.path.join(os.getcwd(),'stocks lists')
file_list = os.listdir(path)

file_list_xlsx = [file for file in file_list if file.endswith('.xlsx')]

for i in file_list_xlsx:
    stocks = i.split("_")
    code = stocks[0]
    name = stocks[1]
    
    df = Update_Stocks_OHCVL(i)
    time.sleep(0.5)
```
